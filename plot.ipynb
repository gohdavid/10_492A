```python
import numpy as np
import seaborn as sns
from utils import plot
import matplotlib.pyplot as plt
```

```python
T = np.linspace(288,320)
A_on = 3.77 * 10**13
A_off = 3.75 * 10**13
E_a_on = 12
E_a_off = 22.06
R = 1.987*10**(-3)
```

```python
ln_of_k_on = np.log(A_on) - E_a_on/R/T
ln_of_k_off = np.log(A_off) - E_a_off/R/T
g = sns.lineplot(x=1/T, y=ln_of_k_on, label="Ab-CRP association")
g = sns.lineplot(x=1/T, y=ln_of_k_off, label="Ab-CRP dissociation")
g.ticklabel_format(style='sci', axis='x', scilimits=(0,1))
g.set_ylabel("Natural log of reaction constant $\ln{k}$")
g.set_xlabel("Inverse temperature $T^{-1}$ [K$^{-1}$]")
g.set_xlim([1/T.max(),1/T.min()])
g.set_title("Linearized Arrhenius plots")
plt.savefig("./figures/kinetics/LinearArrhenius.png")
```

```python
ln_of_k_on = np.log(A_on) - E_a_on/R/T
ln_of_k_off = np.log(A_off) - E_a_off/R/T
g = sns.lineplot(x=T, y=np.exp(ln_of_k_on)/np.exp(ln_of_k_off))
g.set_ylabel("Equilibrium association\nconstant $K_a$ [M$^{-1}$]")
g.set_xlabel("Temperature $T$ [K]")
g.set_xlim([T.min(),T.max()])
g.set_title("Temperature dependent association constant")
plt.savefig("./figures/kinetics/AssociationConstant.png")
```

```python
ln_of_k_on = np.log(A_on) - E_a_on/R/T
ln_of_k_off = np.log(A_off) - E_a_off/R/T
g = sns.lineplot(x=T, y=np.exp(ln_of_k_on), label="$k_{on}$ [s$^{-1}$ M$^{-1}$]")
g = sns.lineplot(x=T, y=np.exp(ln_of_k_off), label="$k_{off}$ [s$^{-1}$]")
g.set_ylabel("Reaction constant")
g.set_xlabel("Temperature $T$ [K]")
g.set_xlim([T.min(),T.max()])
g.set_title("Exponential Arrhenius plots")
plt.savefig("./figures/kinetics/ExpontentialArrhenius.png")
```
